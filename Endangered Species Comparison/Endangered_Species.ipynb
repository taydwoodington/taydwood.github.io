## Transformation of the data

```python
# Define a function to help structure a table and organize the pulled data
def fetch_and_flatten(url, species_type):
    response = requests.get(url)
    data = response.json()
    
    flattened_data = []
    for species in data.get('Species', []):
        species_data = {
            'SpeciesType': species_type,
            'TaxonID': species.get('TaxonID'),
            'ScientificName': species.get('ScientificName'),
            'AcceptedCommonName': species.get('AcceptedCommonName'),
            'KingdomName': species.get('KingdomName'),
            'KingdomCommonName': species.get('KingdomCommonName'),
            'ClassName': species.get('ClassName'),
            'ClassCommonName': species.get('ClassCommonName'),
            'FamilyName': species.get('FamilyName'),
            'FamilyCommonName': species.get('FamilyCommonName'),
            'FamilyRank': species.get('FamilyRank'),
            'IsSuperseded': species.get('IsSuperseded'),
            'ConservationStatus': species.get('ConservationStatus', {}).get('NCAStatus'),
            'Endemicity': species.get('Endemicity'),
            'Synonyms': ', '.join(species.get('Synonyms', [])),
            'PestStatus': species.get('PestStatus'),
            'Species Environment': species.get('SpeciesEnvironment'),  # Fixed key name
            'Species Profile Url': species.get('SpeciesProfileUrl')  # Fixed key name
        }
        flattened_data.append(species_data)
    
    return pd.DataFrame(flattened_data)
```

```python
# URLs for each dataset
urls = {
    'bird': 'https://apps.des.qld.gov.au/species/?op=getspecies&kingdom=animals&family=columbidae',
    'bat': 'https://apps.des.qld.gov.au/species/?op=getspecies&kingdom=animals&family=hipposideridae',
    'treefrog': 'https://apps.des.qld.gov.au/species/?op=getspecies&kingdom=animals&family=hylidae',
    'reptile': 'https://apps.des.qld.gov.au/species/?op=getspecies&kingdom=animals&family=Cheloniidae',
    'snail': 'https://apps.des.qld.gov.au/species/?op=getspecies&kingdom=animals&family=Turbinidae'
}
```
